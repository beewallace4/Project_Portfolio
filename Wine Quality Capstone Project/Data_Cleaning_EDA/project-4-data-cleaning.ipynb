```python
 # Dependencies and Setup
import pandas as pd

# File to Load
white_wine = "resources/white.csv"
red_wine = "resources/red.csv"

# Read red wine and white wine data file and store into Pandas DataFrames
white_df = pd.read_csv(white_wine)
red_df = pd.read_csv(red_wine)
```

```python
white_df.info()
```

```python
red_df.info()
```

```python
white_df["Type"] = "white"
white_df.head()
```

```python
red_df["Type"] = "red"
red_df.head()
```

```python
# Combine the data into a single dataset.  
all_df = pd.concat([white_df, red_df], axis=0).reset_index(drop=True)
```
